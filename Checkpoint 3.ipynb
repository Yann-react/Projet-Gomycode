<h3>VISUALISATION DE LA CORRELATION ENTRE SEX ET AGE</h3>

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map(df_encoder)
```

<p>this function allows you to see the correlation between the different numeric variables of our dataFrame
so we can see that the variable Parch is strongly linked to SibSp which is quite logical plus these two variables
concerns the family and people with a tendency to travel as a family if we use this analysis to see the survival rate between these variables </p>


```python
df[["Survived", "SibSp","Parch"]].groupby(["Survived"], as_index=True).mean()
```

<p>on constate que ceux qui avait des parents a bord ont survecu que les parents eux même</p>

```python
df[["Survived", "Pclass"]].groupby(["Survived"], as_index=True).mean()
```

<p>on constate que les personne ce trouvant dans les classe 2 et 1 ont le plus survecue</p>
