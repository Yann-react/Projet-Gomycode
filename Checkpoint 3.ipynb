```python
df_encoder = pd.concat([df_num,df_cate_encoder],axis=1)
```

<h3>VISUALISATION DE LA CORRELATION ENTRE SEX ET AGE</h3>

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map(df)
```

```python
# cette fonction permet de voir la correlation entre les differents variable numerique de notre dataFrame
# ainsi nous pouvons voir que la variable Parch est fortement lié a SibSp
```
