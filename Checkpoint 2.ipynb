<h1>QUESTION 4</h1>

```python
def missing_char(carac,ind):
    carac = list(carac)
    for i in range(0,len(carac)+1) :
        if i == ind:
            del(carac[i])
            print(carac)
missing_char('kitten',1)
missing_char('kitten',0)
missing_char('kitten',4)
```
